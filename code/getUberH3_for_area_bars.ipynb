```python
import geopandas as gpd
import pandas as pd
```

```python
shape=gpd.read_file(r"nnk_xb.geojson")
```

```python
shape.plot()
```

```python
shape
```

```python
shape_clean=shape.copy()
shape_clean.geometry=shape.buffer(0.0000000001)
```

```python
import h3
```

```python
from tobler.util import h3fy
```

```python
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
dc_hex_res6 = h3fy(shape_clean, resolution=6)
dc_hex_res6 
```

```python
dc_hex_res7 = h3fy(shape_clean, resolution=7)
dc_hex_res7
```

```python
dc_hex_res8 = h3fy(shape_clean, resolution=8)
dc_hex_res8
```

```python
fig, ax = plt.subplots(1, figsize=(15, 15))

dc_hex_res6=dc_hex_res6.to_crs(epsg=3857)
dc_hex_res7=dc_hex_res7.to_crs(epsg=3857)
dc_hex_res8=dc_hex_res8.to_crs(epsg=3857)
shape=shape.to_crs(epsg=3857)


dc_hex_res6.plot(ax=ax, alpha=0.4, linewidth=1.6, edgecolor='white', zorder=1)
dc_hex_res7.plot(ax=ax, alpha=0.4, linewidth=1.6, edgecolor='white')
dc_hex_res8.plot(ax=ax, alpha=0.4, linewidth=1.6, edgecolor='white')
shape.plot(ax=ax, alpha=0.4, linewidth=1.6, edgecolor='black', color='gray')


ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)
ax.axis('off')
```

```python
dc_hex_res9 = h3fy(shape_clean, resolution=9)
dc_hex_res9
```

```python
fig, ax = plt.subplots(1, figsize=(20, 20))

dc_hex_res9=dc_hex_res9.to_crs(epsg=3857)
shape=shape.to_crs(epsg=3857)


dc_hex_res9.plot(ax=ax, alpha=0.4, linewidth=1.6, edgecolor='white')
shape.plot(ax=ax, alpha=0.4, linewidth=1.6, edgecolor='black', color='gray')


ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)
ax.axis('off')
```

```python
dc_hex_res9.to_file('hex_res9_nnk.gpkg', driver='GPKG', layer='name')  
```
